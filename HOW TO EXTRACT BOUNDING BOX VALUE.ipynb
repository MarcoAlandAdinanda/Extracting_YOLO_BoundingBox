# HOW TO EXTRACT BOUNDING BOX VALUE 

## 1. Collecting Samples
### Run this code bellow for a little while and Interrupt the Kernel

```python
from ultralytics import YOLO

model = YOLO("smallyolo.pt")

results = model.predict(source="0", show=True, stream=True)

xyxys = []
confidences = []
class_ids = []


for r in results:
    boxes = r.boxes.cpu().numpy()

    xyxys.append(boxes.xyxy)
    class_ids.append(boxes.cls)
    confidences.append(boxes.conf)

    print(f"class: {class_ids}, xyxys: {xyxys}")
```

## 2. PROCESS THE LISTS 
### NUMBER OF ROWS IS EQUAL TO NUMBER OF OBJECT IN ONE FRAMES

```python
# XYXY FORMAT
print("[[ X_min, Y_min, X_max, Y_max ]]")
print(xyxys[0])
```
